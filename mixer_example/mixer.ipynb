```python
import threading
import time
# это можно юзать для прогрессбара, скорее всего это лучше сделать в music классе, т.е. забей))
class Progressbar:
      
    def __init__(self):
        self._running = True

    def terminate(self):
        self._running = False

    def run(self, n):
        i = 0
        while self._running and i < n:
            print('i :', i)
            i += 1
            time.sleep(1)
```

```python
c = Progressbar()
t = threading.Thread(target = c.run, args =(10, )) #
t.start()
```

```python
# Signal termination
c.terminate() 
  
# Wait for actual termination (if needed) 
t.join() 
```
